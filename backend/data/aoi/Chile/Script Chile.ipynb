```python
import pandas as pd
import geopandas as gpd
```

```python
#regiones = estados
regiones = gpd.read_file("regiones_SIMP.geojson")
```

```python
regiones["id"] = regiones["objectid"]
regiones["name"] = regiones["Region"]
regiones["poly_type"] = "periphery"

regiones = regiones[["id", "name", "poly_type", "geometry"]]
```

```python
regiones["id"] = "PRO-" + regiones["id"].astype(str)
```

```python
regiones.to_file("regiones_SIMP.geojson")
```

```python
#comunas = municipios
comunas = gpd.read_file("comunas.geojson")
```

```python
comunas["id"] = comunas["objectid"]
comunas["name"] = comunas["Comuna"]
comunas["region"] = comunas["Region"]
comunas["poly_type"] = "periphery"
comunas["id"] = "MUN-" + comunas["id"].astype(str)

comunas = comunas[["id", "name", "region", "poly_type", "geometry"]]
```

```python
comunas.to_file("comunas_SIMP.geojson")
```
