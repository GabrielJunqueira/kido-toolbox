```python
import pandas as pd
import geopandas as gpd
```

```python
sp = gpd.read_file("SIMP/SP_Municipios.geojson")
```

```python
iguape = sp[sp["name"] == "Iguape"]
```

```python
iguape["poly_type"] = "core"
```

```python
iguape.to_file("iguape.geojson")
```

```python
cananeia = sp[sp["name"] == "Canan√©ia"]
cananeia["poly_type"] = "core"
cananeia.to_file("cananeia.geojson")
```

```python
iguape_cananeia = pd.concat([iguape, cananeia], ignore_index=True)
```

```python
iguape_cananeia.to_file("iguape_cananeia.geojson")
```
